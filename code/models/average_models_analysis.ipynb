- Graphe 8 - Pie chart des résidus outliers positfs : Ce graphe montre la répartition des modalités des variables qualitatives parmi les résidus outliers positifs, c'est-à-dire pour lesquels le modèle a sous-estimé la valeur du prix de l'hotel.
