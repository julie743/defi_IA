### Analyse des résultats 

**Temps d'entrainement :** 23 secondes

**Score sur l'échantillon de validation :**

- RMSE =  23.094964457851727
- R2 =  0.8935866800494014
- MAE =  15.45985450091411

Le RMSE indique une erreur de 23.09 €, et le MAE une erreur de 15.45 €. 
On remarque que le RMSE est plus élevé que le MAE. Le RMSE est en effet plus fortement impacté par la présence d'outliers que le MAE. C'est-à-dire que, s'il y a quelques points pour lesquels on a fourni une prédiction très éloignée de la réalité, le RMSE y sera plus sensible que le MAE. Quand on trace le graphe des résidus on constet en effet quelques points pour lesquels les résidus ont des valeurs extrêmes (erreur de 150 €)

**Analyse des graphiques :**
- Graphe 1 - observations réelles contre les prédictions : On constate que plus le prix est élevé, plus le modèle fait des erreurs. Dans l'idéal, se graphique doit se rapprocher le plus possible de la droite identité. On constate bien la présence d'une droite se rapprochant de f(x)=x mais quand le prix augmente, le nuage s'élargi en entonnoir. 

- Graphe 2 - résidus contre les valeurs prédites : Ce graphe confirme l'observation précédente, quand le prix à prédire est élevé, le modèle fait plus d'erreur. On conste une forme en entonnoir du nuage de points. Dans l'idéal on devrait avoir une répartition uniforme des points, et que les points soient proches de l'axe 0.

- Graphe 3 - Histogramme des résidus : Ce graphe est utilisé pour voir la répartition des erreurs sur les prédictions. On constate que la majeure partie des erreurs se situent de dans l'intervalle [-30,30] avec un pic à 0. Par ailleurs, il est intéressant de noter qu'il y a davantage de résidus négatifs. Or les résidus sont calculés par : $résidus = valeurs\_réelles - valeurs\_predites$. Donc un résidu négatif signifie que le modèle a surestimé le prix de l'hotel.

**Analyse des graphiques :**
- Graphe 1 - observations réelles contre les prédictions : On constate que plus le prix est élevé, plus le modèle fait des erreurs. Dans l'idéal, se graphique doit se rapprocher le plus possible de la droite identité. On constate bien la présence d'une droite se rapprochant de f(x)=x mais quand le prix augmente, le nuage s'élargi en entonnoir.

Les valeurs des résidus sont plus élevées que pour le modèle linéaire. 

- Graphe 3 - Histogramme des résidus : Ce graphe est utilisé pour voir la répartition des erreurs sur les prédictions. Or les résidus sont calculés par : $résidus = valeurs\_réelles - valeurs\_predites$. Donc un résidu positif signifie que le modèle a sous-estimé le prix de l'hotel. Le modèle de regression linéaire faisait l'inverse, il serait donc peut être intéressant de faire une moyenne des deux prédictions pour obtenir la prédiction finale.
